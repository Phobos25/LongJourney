Здесь мы рассмотрим особенности создания векторов библиотекой Numpy

Основная проблема в том, что, большинство людей, особенно малознакомых с numpy или python неправильно инициализируют вектор. Вместо вектора, они создают структуру или массив первого ранга или одномерный массив. Надо иметь в виду, что одномерный массив не является матрицей. А вектор - это одномерная матрица. 

```python
import numpy as np

#создаем массив из 5 элементов со значениям распределенными 
#по нормальному закону
a = np.random.randn(5)

#смотрим размерность массива
print(a.shape)
'''
размерность выглядит как (5,0)
это означает, что мы создали структуру а не вектор!
'''
```

```python
'''
одной из особенностей является то, что при транспонировании 
сама структура не меняет формы или размерности
'''
print(a)
print(a.T)
```

```python
'''
умножение вектора на его транспонированный вектор, должен давать 
матрицу
'''
print(np.dot(a,a.T))
#но как мы видим, в ответе, получается одно значение.
```

```python
#правильно создавать вектор нужно так, задать ему размерность 1, либо по 
#столбцам либо по строкам
#в данном случае 5 строк и 1 столбец
a = np.random.randn(5,1)
print(a)
print(a.T)

print('\n Произведение вектора на транспонированный вектор:')
#произведение вектора на его транспонированный вектор - матрица!
print(np.dot(a,a.T))
```

```python
#проверяем, является ли a вектором (или матрицей (5,1)). 
assert(a.shape == (5,1))
#assert выдает ошибку, если переменная a не является вектором, 
#либо имеет другую размерность
```
