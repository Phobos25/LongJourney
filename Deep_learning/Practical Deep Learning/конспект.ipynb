Есть специальные сервисы для поиска рисунков, такие как Azure от майкрософта. 1000 запросов в месяц бесплатно. За один запрос можно загрузить около 150 рисунков, что более чем достаточно для обучения модели. Особенно, если используется уже обученная модель и мы дообучаем ее. 

Такие ресурсы периодически появляются и исчезают. Но на данный момент (апрель 2021) Azure все еще работает хорошо.  Поэтому ниже опишу как им пользоваться:

1) Регистрируемся на [Azure Microsoft](https://www.microsoft.com/en-us/bing/apis/bing-web-search-api) --- довольно гемморно регистрироваться, но справиться можно без особых танцов с бубном;

2) Получаем ключ и вводим в jupyter notebook командой: 

```key = os.environ.get('AZURE_SEARCH_KEY', 'your_code')```

3) для поиска используется функция fastai search_images_bing, который требует на вход key и тег, по которым мы будем искать (будем искать панду). код будет выглядеть вот так: 

```result= search_images_bing(key, 'panda')```

4) После этого нам надо взять ссылки на рисунки, т.к. search_images_bing возвращает данные не в самом понятном варианте. На момент записи мы искали значение 'contentUrl' для этого используется код:

```ims = results.attrgot('contentUrl')```

5) Но это только ссылки. Нам надо загрузить сами рисунки на нашу машину. Для этого воспользуемся функцией fastai download_images:

```download_images(dest, urls=result.attrgot('contentUrl'))```
тут dest --- это папка в которую мы загружаем наши рисунки, **надо попробовать оставить пустым, по идее должен записать в корневую папку.** 

* Ниже приводится полный код поиска и загрузки рисунков в разные папки. Папки = метке рисунка:

```
#сохраняем наш код для использования сервиса Azure
key = os.environ.get('AZURE_SEARCH_KEY', 'your_key_here')

# какие рисунки мы будем искать
bear_types = 'grizzly','panda','teddy','polar'
#создаем путь до папки bears, внутри которого будут расположены наши разные типы медведей
path = Path('bears')

if not path.exists():#если папка не существует. Если существует, то нет повода заново загр
    path.mkdir() # создаем папку bears
    for o in bear_types: #идем по каждому виду медведей за раз
        dest = (path/o) #ставим папку загрузки
        dest.mkdir(exist_ok=True) #создаем папку, даже если она уже существует
        results = search_images_bing(key, f'{o} bear') #поиск рисунков
        download_images(dest, urls=results.attrgot('contentUrl')) #загр рис в папку
```

### Data Augmentation / Аугментация данных

Аугментация данных используется в тех случаях, когда данных не достаточно, и надо увеличить их число. Обычно это касается рисунков, т.к. хоть рисунков и много, большая часть не отмечена, не имеет правильых обозначений (labels)

1) Одна из методом это сжатие, когда рсунок сжимается по горизонтали или вертикали. Она имеет существенный недостаток, не существует в природе таких сжатых вещей. Модель будет обучаться на не существующих данных. Но для того, чтобы увеличить число данных и уменьшить оверфит, этот метод достаточен

2) Второй метод это паддинг, когда изображение уменьшается и обрамляется в рамку. Проблема в том, что слишком много пиксеей тратится впустую

3) Зум, когда картинка увеличивается или уменьшатеся. Т.е. как бы дается фокус на определнную часть рисунка. Этот вариант идеален если число данных слишком мало. Но есть и недостатки, уделяется слишком много внимания на несущественные детали. Но такой подход существенно снижает вероятность оверфита

4) Изменение контраста, гаммы, наклона,перспективы и зеркальное отображение. Это самые эффективные и наиболее часто использумые методы аугментации рисунков природы, животных, птиц и т.д. 

в fastai используется 4-й метод, для него есть своя команда ```batch_tfms=aug_transforms``` --- применяет все перечисленные в 4-м методы. Следует еще иметь в виду, что для аугментации используется batch, т.е. аугментируется сразу несколько изображений в видеокарте.

### Создание приложения

После того как мы обучили модель, мы можем сохранить свои веса и архитектуру модели. Это лучше делать постоянно, чтобы не переобучать нашу модель с самого начала. Такая сохраненная модель позволяет использовать обученную модель в приложениях!

Jupyter notebook поддерживает виджеты, которые, по сути, канва в котором будет отображаться рисунок или текст, которые будет меняться в зависимости от переменых. В отличие от обычных функций, они интерактивные и меняются в реальном времени. Т.е. не надо заново запускать эти виджеты. Один раз запустить достаточно. Надо просто создать функцию, которая будет предсказывать и показывать рисунок, и создать event handler, которые при приходе ивента вызывает эту функцию. 

Есть еще расширение voila, которая способдна объединять виджеты в одно большое приложение с кнопками и т.п. Основным преимуществом является то, что для работы не нужен jupyter notebook
