#### Check for errors in images

Let's check whether images are corrupt or not

```python
fns = get_image_files(path)
fns

# search for failed to load images
failed = verify_images(fns)
failed
```

```python
# remove from Path
failed.map(Path.unlink)
```

### Create dataloaders

Dataloaders are class that allows us to keep our train and validation data together in one variable. In addition it allows us to keep dependend and independent variables together as well. resizes images to one size.



```python
ducks = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
dls = ducks.dataloaders(path)
```

```python
# let's look at train dataset
dls.valid.show_batch(max_n=12, nrows=3)
```

### Data augmentation

Since we don't have enough data we can use data augmentation to increase our dataset

```python
ducks = ducks.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())
dls = ducks.dataloaders(path)
```

### Train our model

```python
# let's fine tune and use resnet18
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
# check for errors with confusion matrix
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5, nrows=1)
```

```python
cleaner = ImageClassifierCleaner(learn)
cleaner
```
