```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
```

```python
import gensim.downloader as api
# lets use pretrained model
wv = api.load('word2vec-google-news-300')
vec_king = wv['king']
```

```python
vec_king
```

```python
wv['max']
```

```python
# lets see cosine similarity
wv.most_similar('man')
```

```python
wv.similarity('man', 'king')
```

```python
wv.similarity('programmer', 'html')
```

```python
vec = wv['king'] - wv['man'] + wv['woman']
wv.most_similar([vec])
```
