```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

import statsmodels.api as sm

%matplotlib inline
```

```python
data = pd.read_csv('data/Advertising.csv', index_col=0)
data.head()
```

## Simple linear regression

```python
plt.figure(figsize=(16,8))
plt.scatter(data['TV'], data['sales'], c='black')
plt.xlabel('Money spent on TV ads ($)')
plt.ylabel('Sales (k$)')
```

```python
#данные для обучения
X = data['TV'].values.reshape(-1,1)
#искомый результат
y = data['sales'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(X,y)
#для линейной модели формула будет формулой прямой y = a + b*x
print(f'The linear model is: \n Y = {reg.intercept_[0]} + {reg.coef_[0][0]}*TV')
```

```python
predictions = reg.predict(X)

plt.figure(figsize=(16,8))
plt.scatter(X, y, c='black')
#выведем аппроксимирующую кривую
plt.plot(X, predictions, c='blue', linewidth=2)
plt.xlabel('Money spent on TV ads ($)')
plt.ylabel('Sales (k$)')
```

```python
# проверяем "качество" нашей модели. Или ошибку аппроксимации

X = data['TV']
y = data['sales']

exog = sm.add_constant(X)
est = sm.OLS(y, exog).fit() #OLS - Ordinary Least Squares. Метод наименьших квадратов

print(est.summary())
```

## Multiple linear regression

```python
#мы берем все данные, кроме искомых данных, т.е. таргета y.  
Xs = data.drop(['sales'], axis=1)
y = data['sales'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(Xs, y)

print(f'The linear model is: \n Y = {reg.intercept_[0]} + {reg.coef_[0][0]}*TV + {reg.coef_[0][1]}*radio + {reg.coef_[0][2]}*newspaper')
```

```python
X = np.column_stack((data['TV'], data['radio'], data['newspaper']))
#меняем форму на вектор
y = data['sales'].values.reshape(-1,1)

exog = sm.add_constant(X)
est = sm.OLS(y, exog).fit() #OLS - Ordinary Least Squares. Метод наименьших квадратов

print(est.summary())
```
