```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import sklearn libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
#create or import data
data = make_blobs(n_samples = 500, n_features=2, centers = 5, cluster_std = 2.0, random_state = 101)
X = data[0]
```

```python
#visualize the data
plt.scatter(X[:,0],X[:,1], c=data[1], cmap='rainbow')
```

```python
kmeans = KMeans(5)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
#drawing data with cluster centers 
fig, ax = plt.subplots() 
#plot data with found labels
ax.scatter(X[:,0],X[:,1],c=kmeans.labels_, cmap='rainbow')

# draw circles around cluster centers with radii = radius
radius = 1.0 # circle radius
for i in range(kmeans.n_clusters): 
    circle = plt.Circle((kmeans.cluster_centers_[i,0],kmeans.cluster_centers_[i,1]), radius, color='black', fill=False)
    ax.add_artist(circle)
```

```python
#estimate sum of squared errors (SSE) of K-means model
#we need that to get optimal number of clusters
#Elbow method
X = data[0] # our main data

#number of clusters
k_list = list(range(1,10))

sse = list() 
for k in k_list:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)#train our model
    sse.append(kmeans.inertia_) #SSE of the kmeans model 

# Plot sse against k
plt.figure(figsize=(6, 6))
plt.plot(k_list, sse, '-o')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of squared distance')
plt.title('Elbow method')
```

# as we can see optimal number of clusters is 5, hence it improves little after that. This is so-called Elbow method
