## Невод установка

у невода по данным 2012 по 2018 гг разница между моделью и экспериментом была начиная с энергии $3\times10^{17}$ эВ. 

при таких энергиях плотность потока мюонов соотвествовала более-менее только для самых тяжелых частиц. Что не соответствует другим измерениям ливней, таким как Xmax и его флуктуации. 

В целом для ливней с энергией выше $10^{18}$ эВ в экспериментах наблюдается гораздо больше мюонов чем предсказывает модель. 

**Измерение спектра плотности мюонов это одна из возможных способов решения данной проблемы**, для разных зенитных углов.

для больших зенитных углов, ШАЛ на уровне земли почти полностью состоит из мюонов. Еще его называют мюонным ливнем. Из-за пространственного спреда, который вызывается разлетом частиц в ливне и распдаом мезонов и отклонением магнитным полем Земли, пространственное распределение будет занимать несколько километров. В таких случаях детекторв размером в десятки метров может считаться точечной. 

В индивидуальных событиях измеряется плотность мюонов в точке наблюдения, которое в первом приближении можно считать как отношение числа мюонов к площади детектора $D\sim m/S_{det}$. Распределение событий в плотности мюонов образует локальный спектр плотности мюонов. 


**Необходимо детальное понимание физики адронных взаимодейтсвий в ШАЛ**

Зависимость массового состава от энергии КЛ несет в себе уникальную информацию об источниках и распространеии КЛ.  Еще одним перспективным направлением является поиск источников по совпадению измерений в так называем мульти-мессенджерном методе, или статистической корреляции с потенциальным источниками. 

массовый состав можно измерить по данным Xmax и по числу мюонов $N_{\mu}$. На данный момент эти данные отличаются. 
Для оценки массового состава эти значения сравниваются с моделями, которых много, но на данный момент есть три лучшие это SIBYLL 2.3с, EPOS-LHC, QGSJETII-04, или же пост БАКовские модели, т.к. они были скалиброваны по данным БАКа. 

**Тестирование и измерение свойств адронных взаимодействий** --- еще один метод исследования данного вопроса.. Это также проверка стандартных моделей физики в экстремальных случаях. 

Но надо иметь в виду, что для провеки этих взаимодействий нужно брать самые высокие энергии. Для ШАЛ это энергия в $10^{20}$ эВ или для БАКа (столкновение двух нуклонов) это 432 ТэВ, что в 33 раза выше, чем доступно на данный момент. 

**Измерение эффективного поперечного сечения** для взаимодействия протон-воздух. Это наиболее прямое измерение адронных взаимодействия методом ШАЛ. Измерения основаны на наклоне хвоста распределения Xmax


Рождение мюонов в атмосфере, как зависимость от глубины максимума. Несоответствие между моделью и экспериментов было найдено для ливней выше $10^{18}$ эВ, по данным большинства установок

Затухание потока мюонов на больших углах. Модель предсказывает гораздо более сильное затухание мюонов на больших зенитных углах, по сравнению с экспериментами. 

Нарастание фронта сигнала. Т.к. мюоны приходят в ливне раньше электронов и фотонов, вклад в передний фронт от 10 до 50% общей амплитуды будут вносить именно мюоны. Здесь тоже наблюдатеся несоответствие между моделями и экспериментом. 



### От чего зависит плотность потока мюонов на уровне наблюдения

1) От энергии ливня

2) От зенитного угла

3) Возраста ливня (это такая величина, которая зависит от высоты над уровнем моря, атмосферы и зенитного угла прихода)

4) пространственное расстояние r от оси ливня

5) Порог энергий для детекторов мюонов

Определение первичной массы, направления прихода и энергии космических лучей, приходящих на Землю с энергией, превышающей 1015 эВ, осуществляется посредством обнаружения и анализа широких атмосферных ливней (ШАЛ). Двумя наиболее часто используемыми наблюдаемыми для вывода средней логарифмической массы основного потока, 〈lnA〉, являются глубина максимума ливня в атмосфере Xmax и количество мюонов в ливне Nμ. Один из основных источников систематики в 〈lnA〉 проистекает из неопределенности в моделях взаимодействия адронов высоких энергий, используемых при моделировании атмосферных ливней. В то время как Xmax связан с электромагнитной составляющей ШАЛ, Nμ напрямую проистекает из адронного каскада, который более подвержен неопределенностям модели. Проверяя согласованность Xmax и Nμ с точки зрения интерпретации массы, модели могут быть протестированы, открывая окно для физики частиц в областях фазового пространства за пределами досягаемости LHC. В то же время модели могут быть улучшены, что позволяет более точно определять массу основных цветов.

Изучение свойств широких атмосферных ливней служит с одной стороны, базой для изучения ядерных взаимодействий в области СВЭ, которые еще не достигнуты на современных ускорителях, с другой стороны, для изучения характеристик первичных космических лучей – энергетического спектра распределения направлений прихода на небесной сфере, массовый состав и природа космических лучей. Все эти характеристики очень важны для решения задачи происхождения космических лучей, т.е. выявления основных источников, механизма ускорения и процессов распространения в космической среде.
