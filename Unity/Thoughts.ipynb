## Update 2021/04/15

Последний раз я занимался юнити в 2018 году летом. В начале того года, я решил всерьез заняться разработкой компьютерной игры. Я чувствовал, что не до конца понимаю как это делается от начала до конца, поэтому купил курсы юнити на udemy. Через полгода я бросил это дело. 

Прошло три года, и теперь я снова начал заниматься юнити. Я не буду слишком сильно давить на себя, буду просто делать игры для себя. Чтобы разобраться с юнити, присоединиться к какому-нибудь коммюнити. Поучаствовать в ludum dare. Это моя цель. 

Я понял одну важную вещь, за эти три года, которые я ничего не делал. Чтобы разобраться как делать компьютерную игру надо от начала до конца сделать какую-либо игру. Для этого берем игру, любую, пусть старую. Скажем, пакмен или герои 2/3, мастер оф орион. И просто воссоздаем эту игру. Это называется backward engineering. Так я понимаю как это они сделали, UI, геймплей. При этом можно что-то улучшить или поменять, что было реализовано не очень хорошо.
